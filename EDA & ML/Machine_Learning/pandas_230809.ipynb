## 열과 행 선택하기

```python
result = df_boston[['CRIM', 'ZN', 'INDUS']]
result.head()
```

```python
#열을 선택할 때, 컬럼명을 따로 정리해주는게 편함
cols = ['CRIM', 'ZN', 'INDUS']
result = df_boston[cols]
result.head(1)
```

```python
# ZN이 18.0에 해당하는 행만 추출
# 행추출의 기본 원리는 조건문
result = df_boston[df_boston['ZN'] == 18.0]
result
```

```python
# AGE가 50 이상
# reset_index(drop=True) 같이 해줘야, 깔끔하게 볼 수 있음
result = df_boston[df_boston['AGE'] >= 50].reset_index(drop=True)
result
```

```python
# INDUS가 5.0이상이면서, B가 394이상인 데이터 (동시만족 and)
# 소괄호로 조건을 묶어주고, & 기호로 연결
# result = df[(조건식1) & (조건식2)] 처럼 틀을 잡아주고 시작

result = df_boston[(df_boston['INDUS'] >= 5.0) & (df_boston['B'] >= 394)].reset_index(drop=True)
result
```

```python
# 또는(or)로 할 경우 : | 기호로 연결
result = df_boston[(df_boston['INDUS'] >= 5.0) | (df_boston['B'] >= 394)].reset_index(drop=True)
result
```

## iloc 와 loc
- iloc (Integer Location-based Indexing): 정수 인덱스를 기반으로 행과 열을 선택
- loc (Label-based Indexing): 레이블을 기반으로 행과 열을 선택 -> 행과 열의 이름(레이블)을 사용하여 데이터에 접근

```python
df_boston.head()
```

```python
#iloc 활용
result = df_boston.iloc[0:3,0:3]
result
```

```python
#loc 활용 >> iloc보다 시간이 좀 더 걸리지만, 데이터 가공하긴 이게 편함
result = df_boston.loc[[0, 1, 2],['CRIM', 'ZN', 'INDUS']]
result = df_boston.loc[0:2, ['CRIM', 'ZN', 'INDUS']]

# 이렇게 조건을 걸어서 행열 추출시 loc가 더 유용
result2 = df_boston.loc[df_boston['INDUS'] >= 5.0,['CRIM', 'ZN', 'INDUS']]
result2
```

```python
# 행열 추출시, 행 추출 조건 2개 이상도 가능
result3 = df_boston.loc[(df_boston['INDUS'] >= 5.0) & (df_boston['B'] >= 394)
                        , ['CRIM', 'ZN', 'INDUS']].reset_index(drop=True)
result3
```

## 결측치 다루기

```python
import pandas as pd
import numpy as np

np.random.seed(0)
data = np.random.randint(0, 10, size = (10, 3)).astype(float)
print(data)
```

```python
data[data < 3] = np.nan
print(data)
```

```python
df = pd.DataFrame(data, columns=["col1", "col2", "col3"]) #외부에서 가져온 새로운 데이터라고 가정
df.info() #결측치가 얼마나 되는지 대충 확인
```

```python
#전체적인 결측치 규모 확인
df.isna().sum()
```

```python
#결측치를 평균으로 대치
col2_mean = df['col2'].mean()
col3_mean = df['col3'].mean()

df['col3'] = df['col3'].fillna(col3_mean)
df['col2'] = df['col2'].fillna(df['col2'].mean())
df
```

```python
df = pd.DataFrame({
    "이름": ["Evan", "Sara", "Chris", "Messi", "Woods", "Jordan"], 
    "국어": [70, 90, None, None, 100, 85], 
    "영어": [65, 100, None, 90, 70, 80]
})

df
```

## 행 추가

```python
df.loc[6, "이름":"영어"] = ['KIM', 50, 60]
df
```

```python
# 국어점수가 90점 이상이면 "A"
# 국어점수가 80점~89점 "B"
# 국어점수가 70점~79점 "C"
# 나머지는 F

def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'F'

df['국어 등급'] = df['국어'].apply(grade)

print(df)
```

### np 조건문
- np.select : 다중조건식, 다중결과값, default => else 조건
- np.where(조건식, 참, np.where(조건식, 참)...)

```python
# numpy와 pandas 연계 예시

condition_list = [(df["국어"] >= 90), (df['국어'] >= 80) & (df['국어'] < 90), (df['국어'] >= 70) & (df['국어'] < 80)]
choice_list = ["A", "B", "C"]
df['점수'] = np.select(condition_list, choice_list, default = "F")
df
```

## shift 기능
- 시계열 데이터 다룰 때 유용
- 예) 전일 대비 증가액/증가율 등

```python
temp_df = sales.loc[:, ['datesold', 'price']].copy()
temp_df.head(3)
```

```python
temp_df['shifted_v1'] = temp_df['price'].shift(1, fill_value=0).astype(int)
temp_df['diff'] = temp_df['price'] - temp_df['shifted_v1']
temp_df.head()
```

```python
temp_df['shifted_v2'] = temp_df['price'].shift(2, fill_value=0).astype(int)
temp_df['diff_2'] = temp_df['price'] - temp_df['shifted_v2']
temp_df.head(1)
```

```python
temp_df = sales.loc[:, ['datesold', 'price']].copy()
temp_df['shifted_v3'] = temp_df['price'].shift(-1, fill_value=0).astype(int)
temp_df['shifted_v4'] = temp_df['price'].shift(-2, fill_value=0).astype(int)

temp_df.tail()
```

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly['total_cases_q'] = pd.qcut(covidtotalsonly['total_cases'],
                                         labels = ['매우낮음', '낮음', '보통', '높음', '매우높음'],
                                         q=5, precision=0)
covidtotalsonly['total_deaths_q'] = pd.qcut(covidtotalsonly['total_deaths'],
                                         labels = ['매우낮음', '낮음', '보통', '높음', '매우높음'],
                                         q=5, precision=0)

covidtotalsonly.head()
```
