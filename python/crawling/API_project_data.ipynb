    ※ 영세자영업 : 개인사업자이면서 5인 미만 100대 생활밀접업종_외식업/소매업/서비스업)을 운영하는 자영업
- 작성방법 : 서울시 상가업소 정보(운영점포수, 개업수, 폐업수, 종업원수, 매출액 등) 및 통계청 SBR(기업통계등록부)자료를 활용하여 지역별 추정

    - 경영활동현황 총괄(업종별, 자치구별)
    - 운영 점포수 현황(종사자 규모별, 자치구별)
    - 개업점포수 현황(종사자 규모별, 자치구별)
    - 폐업점포수 현황(종사자 규모별, 자치구별)
    - 평균영업기간 현황(종사자 규모별, 자치구별 운영점포수)
    - 매출액 현황(종사자 규모별, 종사자 규모별 면적당, 자치구별, 자치구별 면적당, 업종별 면적당) 등
