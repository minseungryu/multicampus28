### XML 데이터를 데이터프레임으로

```python

xml_url = "http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key=test&type=xml&iStartUnitCode=101&iEndUnitCode=103&numOfRows=10&pageNo=1"

req = requests.get(xml_url)
print(req.text)
```

```python
import xml.etree.ElementTree as et

def parse_xml_to_df(xml_data):
    tree = et.ElementTree(et.fromstring(xml_data))

    data_list = []
    root = tree.getroot()
    for item in root.findall('.//realUnitTrtmVO'):
        efcv_trfl = int(item.find('efcvTrfl').text)
        end_unit_code = int(item.find('endUnitCode').text)
        end_unit_name = item.find('endUnitNm').text
        start_end_std_type_code = int(item.find('startEndStdTypeCode').text)
        start_end_std_type_name = item.find('startEndStdTypeNm').text
        start_unit_code = int(item.find('startUnitCode').text)
        start_unit_name = item.find('startUnitNm').text
        std_date = item.find('stdDate').text
        std_time = item.find('stdTime').text
        tcs_car_type_code = int(item.find('tcsCarTypeCode').text)
        tcs_car_type_div_code = int(item.find('tcsCarTypeDivCode').text)
        tcs_car_type_div_name = item.find('tcsCarTypeDivName').text
        tcs_car_type_name = item.find('tcsCarTypeName').text
        time_avg = float(item.find('timeAvg').text)
        time_max = float(item.find('timeMax').text)
        time_min = float(item.find('timeMin').text)
        data_list.append([
            efcv_trfl, end_unit_code, end_unit_name, start_end_std_type_code,
            start_end_std_type_name, start_unit_code, start_unit_name, std_date,
            std_time, tcs_car_type_code, tcs_car_type_div_code, tcs_car_type_div_name,
            tcs_car_type_name, time_avg, time_max, time_min
        ])
        
        columns = [
        'EfcvTrfl', 'EndUnitCode', 'EndUnitName', 'StartEndStdTypeCode',
        'StartEndStdTypeName', 'StartUnitCode', 'StartUnitName', 'StdDate',
        'StdTime', 'TcsCarTypeCode', 'TcsCarTypeDivCode', 'TcsCarTypeDivName',
        'TcsCarTypeName', 'TimeAvg', 'TimeMax', 'TimeMin'
        ]
    df = pd.DataFrame(data_list, columns=columns)
    return df
    
xml_data = req.text
df_xml = parse_xml_to_df(xml_data)
print(df_xml)
```

### xml >> dataframe 변환 (수업코드)

```python

import requests
import lxml
from bs4 import BeautifulSoup

xml_url = "http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key=test&type=xml&iStartUnitCode=101&iEndUnitCode=103&numOfRows=10&pageNo=1"

req = requests.get(xml_url)
#print(req.text)

soup = BeautifulSoup(req.text, "lxml")
#print(soup.prettify())

stddate_list = soup.find_all("stddate")
timeAvg_list = soup.find_all("timeavg")


stddate = []
avgTime = []
for date, timeAvg in zip(stddate_list, timeAvg_list):
    #print(date.get_text(), timeAvg.get_text())
    stddate.append(date.get_text())
    avgTime.append(timeAvg.get_text())
df = pd.DataFrame({'date':stddate, 'avgTime':avgTime})
df
```
