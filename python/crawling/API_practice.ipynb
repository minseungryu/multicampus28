# 서울 열린데이터광장

### xml 파일 데이터프레임으로 (방법1)

```python
import pandas as pd
import xml.etree.ElementTree as ET
import requests
url = "http://openapi.seoul.go.kr:8088/566468557572733034336b65477a52/xml/tbLnOpendataRtmsV/1/5/"
response = requests.get(url)

df = pd.read_xml(response.content)
select_columns = ['ACC_YEAR', 'SGG_NM', 'BJDONG_NM', 'BLDG_NM','OBJ_AMT']
df = df[select_columns]

df.columns = ['연도', '구이름', '동이름', '건물명', '가격']
df = df.dropna().reset_index(drop=True)

df
```

### xml 파일 데이터프레임으로 (방법2)

```python
import pandas as pd
import xml.etree.ElementTree as ET
import requests

url = "http://openapi.seoul.go.kr:8088/566468557572733034336b65477a52/xml/tbLnOpendataRtmsV/1/5/"
response = requests.get(url)

# XML 데이터 파싱
root = ET.fromstring(response.content)

# 필요한 컬럼 선택하고 데이터프레임으로 변환
data_list = []
for item in root.findall('./row'):
    data = {}
    data['연도'] = int(item.find('ACC_YEAR').text)
    data['구이름'] = item.find('SGG_NM').text
    data['동이름'] = item.find('BJDONG_NM').text
    data['건물명'] = item.find('BLDG_NM').text
    data['가격'] = int(item.find('OBJ_AMT').text)
    data_list.append(data)

df = pd.DataFrame(data_list)

print(df)
```
