### 상권 센터포인트

### 테스트 코드

```python
# 중심점(상권 중앙정보) => 리스트로
center_points = (37.496837, 127.028104)

# 반경 몇 미터(meters)/킬로미터(kilometers)인지 지정
radius = 300

# 반경 내에 있는 지점의 위치 정보 list
points_within_radius = []

# 위치를 확인할 데이터(버스정류장, 지하철역 - 위도+경도)
points = [(37.49813573, 127.0291478),(37.4988294, 127.0298391), (37.50235998, 127.0257902)]

for point in points:
    distance = geodesic(center_points, point).meters
    
    if distance <= radius:
        points_within_radius.append(point)

print('-----------반경 내에 있는 지점-------------------')
for point in points_within_radius:
    print(f'[위도]{point[0]}, [경도]{point[1]}')
```

```python
print(market_points_list[:3])
print(sub_position_list[:3])
print(bus_position_list[:3])
```

```python
# 상권별로 n미터내에 몇개의 지하철역이 있는가?
# 결과 : 00상권 ---- 지하철역 00개

# 상권코드 중앙
centerpoints = market_points_list

#상권 반경 00미터
radius = 500

# 반경별 리스트
# bus_within_radius = []
# subway_within_radius = []

subway_stations = sub_position_list
bus_stations = bus_position_list

result = {}

for center in centerpoints:
    bus_count = 0
    subway_count = 0
    
    #버스정류장
    for bus in bus_stations:
        distance = geodesic(center, bus).meters
        if distance <= radius:
            bus_count += 1

    
    #지하철역
    for subway in subway_stations:
        distance = geodesic(center, subway).meters
        if distance <= radius:
            subway_count += 1
    
    #결과를 딕셔너리에 저장
    result[center] = [subway_count, bus_count]

# 결과 출력
for center, counts in result.items():
    print(f'{center} 상권 ---- 지하철역 {counts[0]}개, 버스 정류장 {counts[1]}개')
```
