### y_station에 저장 > 연도에서 분기 추출

```python
y_station.head(2)
```

```python
y_station['분기'] = y_station['월'].apply(lambda x: (int(x) - 1)//3 + 1)
```

### 시간대 그루핑

```python
y_station.columns
```

```python
cols = y_station.columns.tolist()
cols = cols[-1:] + cols[:-1]

y_station = y_station[cols]
y_station
```

```python
y_station.columns
```

```python
y_station['00-06시'] = y_station.loc[:, 
                                    ['00시-01시 승하차','01시-02시 승하차',
       '02시-03시 승하차','03시-04시 승하차','04시-05시 승하차','05시-06시 승하차']].sum(axis = 1)


y_station['06-11시'] = y_station.loc[:, '06시-07시 승하차':'10시-11시 승하차'].sum(axis = 1)
y_station['11-14시'] = y_station.loc[:, '11시-12시 승하차':'13시-14시 승하차'].sum(axis = 1)

y_station['14-17시'] = y_station.loc[:, '14시-15시 승하차':'16시-17시 승하차'].sum(axis = 1)
y_station['17-21시'] = y_station.loc[:, '17시-18시 승하차':'20시-21시 승하차'].sum(axis = 1)
y_station['21-24시'] = y_station.loc[:, '21시-22시 승하차':'23시-24시 승하차'].sum(axis = 1)
```

```python
time_cols = ['04시-05시 승하차', '05시-06시 승하차',
       '06시-07시 승하차', '07시-08시 승하차', '08시-09시 승하차', '09시-10시 승하차',
       '10시-11시 승하차', '11시-12시 승하차', '12시-13시 승하차', '13시-14시 승하차',
       '14시-15시 승하차', '15시-16시 승하차', '16시-17시 승하차', '17시-18시 승하차',
       '18시-19시 승하차', '19시-20시 승하차', '20시-21시 승하차', '21시-22시 승하차',
       '22시-23시 승하차', '23시-24시 승하차', '00시-01시 승하차', '01시-02시 승하차',
       '02시-03시 승하차', '03시-04시 승하차']

final_df = y_station.drop(columns=time_cols, axis = 1)
```

```python
final_df.head(1).T
```

```python
final_df.columns
```

```python
final_cols = ['연도','분기','월','지하철역', '호선명', '00-06시', '06-11시', '11-14시', '14-17시',
       '17-21시', '21-24시']
final_df = final_df[final_cols]
```

```python
final_df2 = final_df.sort_values(by = ['연도', '분기', '월'])
final_df2
```
